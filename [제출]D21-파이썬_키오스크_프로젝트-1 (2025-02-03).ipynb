- **④ 데코레이터로 주문표 양식 출력**

````python
import datetime

def receipt_decorator(func):
    def wrapper(*args, **kwargs):
        print('⟝' + '-'*30 + '⟞')
        for _ in range(5):
            print('|' + ' '*32 + '|')
        result = func(*args, **kwargs)
        for _ in range(5):
            print('|' + ' '*32 + '|')
        print('⟝' + '-'*30 + '⟞')
        return result
    return wrapper

class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    def menu_select(self):
        self.order_menu = []
        self.order_price = []
        self.price_sum = 0

        n = 0
        count = 0
        while n < 1 or n > len(self.menu):
            n = int(input("음료 번호를 입력하세요 : "))

            if 1 <= n <= len(self.menu):
                self.order_price.append(self.price[n-1])
                self.price_sum += self.price[n-1]
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        t = 0
        while t != 1 and t != 2:
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1:
                self.temp = 'HOT'
            elif t == 2:
                self.temp = 'ICE'
            else:
                print("1과 2 중 하나를 입력하세요.\n")

        count += 1
        self.order_menu.append(self.temp + ' ' + self.menu[n-1])
        print('[', count, ']', self.temp, self.menu[n-1], ' : ', self.price_sum, '원')

        while n != 0:
            print()
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))
            if 1 <= n <= len(self.menu):
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = 'HOT'
                    elif t == 2:
                        self.temp = 'ICE'
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                count += 1
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                self.order_price.append(self.price[n-1])
                self.price_sum += self.price[n-1]
                print('[', count, ']', '추가 주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원\n', '합계 : ', self.price_sum, '원')
            elif n == 0:
                print("주문이 완료되었습니다.")
                print(self.order_menu, self.order_price)
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    def pay(self):
        print(f'총 합계금액은 {self.price_sum}원 입니다.')
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input('\n현금 결제는 1번, 카드 결제는 2번을 입력: '))
            if self.pay_method == 1:
                print('직원을 호출하겠습니다.')
            elif self.pay_method == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('다시 결제를 시도해 주세요.')

    @receipt_decorator
    def table(self):
        now = datetime.datetime.now()
        form = f'주문 일시: {now.strftime("%Y-%m-%d %H:%M:%S")}'
        print('| ', end='')
        print(form, '|')

        for i in range(len(self.order_menu)):
            output = '| ' + str(self.order_menu[i]) + ' : ' + str(self.order_price[i]) + '원'
            outline3 = ' ' * (32 - len(output)) + '|'
            print(output + outline3)
        output = '| * 합계 금액 : ' + str(self.price_sum) + '원'
        outline3 = ' ' * (28 - len(output)) + '|'
        print(output + outline3)
````

````python
a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()
````
